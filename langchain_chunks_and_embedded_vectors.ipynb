```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load your text file
with open("/content/data.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,  # Size of each chunk
    chunk_overlap=50  # Overlap between chunks to maintain context
)

# Split the text into chunks
chunks = text_splitter.split_text(text)

# Print or process the chunks
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:")
    print(chunk)
    print("\n" + "="*50 + "\n")

print(f"Total number of chunks: {len(chunks)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer

# Step 1: Load your text file
with open("/content/data.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Step 2: Initialize the text splitter from LangChain
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,  # Maximum characters in each chunk
    chunk_overlap=50  # Overlap between chunks to maintain context
)

# Split the text into chunks
chunks = text_splitter.split_text(text)

# Step 3: Load the pre-trained SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')  # You can choose any model you prefer

# Step 4: Convert each chunk into an embedding vector
chunk_vectors = []
for chunk in chunks:
    vector = model.encode(chunk)
    chunk_vectors.append(vector)
    print(len(vector))

# Step 5: (Optional) Print the vectors
for i, vector in enumerate(chunk_vectors):
    print(f"Chunk {i+1} Vector: {vector[:10]}...")

print(f"Total number of vectors: {len(chunk_vectors)}")
```
